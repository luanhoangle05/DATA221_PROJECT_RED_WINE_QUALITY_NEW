```
# 06_final_comparison.ipynb

# imports
import pandas as pd
import matplotlib.pyplot as plt

# load all model result csv files
logistic_results = pd.read_csv("../output/logistic_regression_results.csv")
knn_results = pd.read_csv("../output/knn_results.csv")
decision_tree_results = pd.read_csv("../output/decision_tree_results.csv")
random_forest_results = pd.read_csv("../output/random_forest_result.csv")
neural_network_results = pd.read_csv("../output/neural_net_result.csv")

# combine all model results into one dataframe
all_results = pd.concat([
    logistic_results,
    knn_results,
    decision_tree_results,
    random_forest_results,
    neural_network_results
], ignore_index=True)

# display combined table
print("Combined Model Comparison Table:")
print(all_results)

# save combined comparison csv
all_results.to_csv("../output/final_model_comparison.csv", index=False)

# set model names as index for easier plotting
comparison_df = all_results.set_index("Model")

# plot accuracy comparison
comparison_df["Accuracy"].plot(kind="bar")
plt.title("Accuracy Comparison")
plt.ylabel("Accuracy")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# plot precision comparison
comparison_df["Precision"].plot(kind="bar")
plt.title("Precision Comparison")
plt.ylabel("Precision")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# plot recall comparison
comparison_df["Recall"].plot(kind="bar")
plt.title("Recall Comparison")
plt.ylabel("Recall")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# plot f1-score comparison
comparison_df["F1-Score"].plot(kind="bar")
plt.title("F1-Score Comparison")
plt.ylabel("F1-Score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# plot roc-auc comparison
comparison_df["ROC-AUC"].plot(kind="bar")
plt.title("ROC-AUC Comparison")
plt.ylabel("ROC-AUC")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# find best model based on F1-score
best_model = comparison_df["F1-Score"].idxmax()
best_f1 = comparison_df["F1-Score"].max()

print("\nBest Model Based on F1-Score:")
print(best_model)
print("F1-Score:", best_f1)
```
