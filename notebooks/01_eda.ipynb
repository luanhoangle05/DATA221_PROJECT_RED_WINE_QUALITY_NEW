```python
'''
importing and loading dataset
'''
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data_of_red_wine = pd.read_csv('../data/winequality-red.csv')
data_of_red_wine.head()
```

```python
'''
summary statistics and all missing values in dataset
'''
display(data_of_red_wine.describe())
print("missing values per column shown...:\n", data_of_red_wine.isnull().sum())
```

```python
'''
featuring all distributions
- histograms for each feature to understand their distribution
'''
data_of_red_wine.hist(bins=15, figsize=(15,10),color='blue',edgecolor='black')
plt.tight_layout()
plt.show()
```

```python
'''
correlation heatmaps
- be able to see correlations between features to be able to see which feature contributes the most to red wine
  quality.
'''
correlation_of_features = data_of_red_wine.corr()
plt.figure(figsize=(12,8))
plt.imshow(correlation_of_features, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(correlation_of_features)), correlation_of_features.columns, rotation=90)
plt.yticks(range(len(correlation_of_features)), correlation_of_features.columns)
plt.title("correlation heatmap")
plt.show()
```
