```python
'''
importing and loading dataset
'''
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# loads X and y train and make it as one whole dataframe with concat (concatenate)
X = pd.read_csv('../output/x_train.csv')
y = pd.read_csv('../output/y_train.csv')
data_of_red_wine = pd.concat([X,y],axis=1)
data_of_red_wine.head()
```

```python
'''
summary statistics and all missing values in dataset
'''
display(data_of_red_wine.describe())
print("missing values per column shown...:\n", data_of_red_wine.isnull().sum())
```

```python
'''
featuring all distributions
- histograms for each feature to understand their distribution
'''
data_of_red_wine.hist(bins=15, figsize=(15,10),color='blue',edgecolor='black')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,10))

correlation_matrix = data_of_red_wine.corr()

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Correlation Heatmap of Wine Features')
plt.show()
```
